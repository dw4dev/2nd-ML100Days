# 作業
* 在精簡深度學習的方式上 : 卷積類神經 (CNN) 採用像素遠近，而遞歸類神經 (RNN) 採用著則是時間遠近
* 那麼，既然有著類似的設計精神，兩者是否有可能互換應用呢?

## ANS:

1. 卷積類神經(CNN)特性為空間擴展性質，神經元與特徵卷積，用於靜態輸出。特點是善於抽取位置不變特徵。<br>
0. 循環神經網絡(RNN)為時間擴展性質，神經元與多個時間輸出計算。可以用於描述時間上連續狀態的輸出，有記憶功能。<br>
RNN可以看成一個在時間上傳遞的神經網絡，它的深度是時間的長度！

> 相同點：<br>
1. 傳統神經網絡的擴展。
0. 前向計算產生結果，反向計算模型更新。
0. 每層神經網絡橫向可以多個神經元共存,縱向可以有多層神經網絡連接。


> 不同點<br>
1. CNN空間擴展，神經元與特征卷積；RNN時間擴展，神經元與多個時間輸出計算
0. RNN可以用於描述時間上連續狀態的輸出，有記憶功能，CNN用於靜態輸出
0. CNN高級100+深度，RNN深度有限
