```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
"""
Your Code Here
"""
matrix = np.random.randint(-10,10,size=(10,10))/10

plt.figure(figsize=(10,10))
"""
Your Code Here
"""
heatmap = sns.heatmap(matrix)

plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
matrix = np.random.randint(-10,10,size=(nrow,ncol))/10

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(plt.hist)
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```

```python
nrow = 1000
ncol = 3
"""
Your Code Here
"""
'''
numpy.random.normal(loc=0.0, scale=1.0, size=None)
產生符合常態分配的亂數
    loc: Mean (“centre”) of the distribution. 中央
    scale: Standard deviation (spread or “width”) of the distribution. 分布標準差，即圖形的寬度。
    size: int or tuple of ints, Output shape. If the given shape is, 
          e.g., (m, n, k), then m * n * k samples are drawn.
'''
matrix = np.random.normal(1,0.4,size=(nrow,ncol))/10

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
"""
Your Code Here
Please replace "..." to correct plot function
"""
grid.map_upper(plt.scatter , alpha = 0.2)
grid.map_diag(plt.hist )
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r)

plt.show()
```

```python
# 常態分佈圖形
norarray = np.random.normal(1,0.5,size=nrow*ncol)/10
sns.kdeplot(norarray).set_title('Normal Distribution with width 0.5 (from -0.1 ~ 0.3)')
```
