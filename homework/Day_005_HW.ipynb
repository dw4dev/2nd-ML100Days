```python
'''
使用 pandas+matplotlib 繪製直方圖
bins:
    Number of histogram bins to be used. If an integer is given, 
    bins + 1 bin edges are calculated and returned. 
    If bins is a sequence, gives bin edges, 
    including left edge of first bin and right edge of last bin. 
    In this case, bins is returned unmodified.
'''
app_train['AMT_ANNUITY'].plot(kind='hist', bins=30)
# 等效語法
# app_train['AMT_ANNUITY'].plot.hist(bins=30)
```
