#### 資料的 row 數以及 column 數

```python
'''
方法1
app_train.columns       # 會回傳所有欄位名稱(索引名稱、項目)
app_train.index         # 會回傳所有列的索引項目，是一個 <class 'pandas.core.indexes.range.RangeIndex'> 物件
'''
print('column count:', len(app_train.columns))
print('row count:', len(app_train.index))
```

```python
'''
方法2
透過使用 app_train.shape 方法來取得資料表格的尺寸(tuple)資訊，
'''
(rows, cols) = app_train.shape

print('column count:', cols)
print('row count:', rows)
```

#### 列出所有欄位

```python
col_width = len(max(app_train.columns, key=len))
print('max column name length:', col_width, '\n')

cc = 0; item_per_line = 4;
colstr=""
for c in app_train.columns:
    colstr += "{:<30}".format(c)
    cc+=1
    if(cc % item_per_line == 0):
        print(colstr)
        colstr=""
```

```python
'''
column 也是一個 Series
'''
cidx = app_train.columns     # 取得行索引物件
        # app_train.columns.tolist()           # 將資料轉換為 list
ridx = app_train.index       # 取得列索引物件

# 取得指定欄位序列值
app_train.AMT_INCOME_TOTAL      # 把欄位名稱當作屬性使用，取得該欄位的所有列值，即 取得該欄的 序列值(Serial)
# app_train['AMT_INCOME_TOTAL']   # 等價使用方式
```

```python
# 欄位名稱變更
app_train.rename(columns={'SK_ID_CURR':'SKID'}, inplace=True)

# 取得指定欄號的名稱，即欄位名稱
print('new column name:', app_train.columns[0])          # 會顯示修改後的新名稱

# 取得指定欄位名稱的序列資料
print(app_train.SKID)
```

```python
'''
計算所有欄位中，資料序列非 null, `None`, `NaN`, `NaT` 的個數
參數 axis 為統計方式， 0 以欄統計(預設)；1 以列來統計
''' 
print(app_train.count())        # 列出所有欄位的統計資料

print('-'*120)

# 指定取得指定的欄位資料個數值
print('column "AMT_REQ_CREDIT_BUREAU_HOUR" data count:', app_train['AMT_REQ_CREDIT_BUREAU_HOUR'].count())
```

```python
'''
資料表操作
'''

print(app_train.iloc[3])                   # 取得指定索引號的列資料

new_app_train = app_train.drop(3)          # 刪除指定索引的列資料
                                           #   然後會回傳一個新的 DataFrame
print('-'*120)
print(new_app_train.iloc[3])
```
