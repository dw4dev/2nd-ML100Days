## ANS:

> 1. In a convolutional neural network, there are 3 main parameters that need to be tweaked to modify the behavior of a convolutional layer.<br>
These parameters are (1)filter size, (2)stride and (3)zero padding. <br>
The size of the output feature map generated depends on the above 3 important parameters.<br>
公式：<br><br>
當 filter size (K) =  $3 × 3$ 時，<br>
input size $= 256 × 256$<br>
 $stride =\frac{K-1}{2} $，where K is the filter size.  計算後會得到 $stride(S) = 2$<br>
 Padding Size $ = 1 × 1 $<br>
帶入公式後：$ \lfloor \frac{256+2*1-3}{2} \rfloor + 1=128 $.<br>
最後會得到 $ 128(Width) × 128(height) × 128(filters) $ 張特徵圖  (feature maps)。<br>
